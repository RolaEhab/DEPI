- I want the bank to have a list of customers with name and id and balance
- so i enter the bank 
- state if i want to register or login 
- if register take hime to registration and add him to the list then proceed with the login 
- if login take him to the options of 
- Main Menu (Loop):
1: Check balance → show balance.
2: Deposit → ask amount, add to balance.
3: Withdraw → ask amount, subtract if enough balance, else warn.
4: Exit → end program with a thank-you message.

```python
class Bank : 
    def __init__(self):
        self.customers = []
        self.curr_user = None
        self.curr_balance = None
     # Checks if a numper is negative or invalid to return none and a print statment   
    def num_check(self, input):
        try:
            num = int(input)
            if num < 0:
                print(f"You Entered a negative number please enter a valid number.")
                return None
            return num
        except ValueError:
            print(f"You Entered an invalid input please enter a valid number.")
            return None
        
    def register(self, name, pin, balance):
        if self.num_check(balance) and self.num_check(pin):
            balance = int(balance)
            self.customers.append({"name": name, "pin" : pin , "balance" : balance})
            print(f"Congrats {name} you registered successfuly!")
            
    def login(self, name, pin):
        for cust in self.customers:
            if(self.num_check(pin) and cust["name"] == name and cust["pin"] == pin):
                self.curr_user = name
                self.curr_balance = cust['balance']
                print(f"Welcome again {self.curr_user} your current balance is {self.curr_balance} EGP.")
                return True
            else:
                return False
        
    def check_balance(self):
        print(f"{self.curr_user} you currently have {self.curr_balance} EGP.")
        
    def deposit(self, amount):
        amount = self.num_check(amount)
        if(amount is not None):
            self.curr_balance = self.curr_balance + amount
            print(f"Thank you {self.curr_user} for depositing with us right now you have {self.curr_balance} EGP in your acount.")
            # print(f"data type of balance : {self.curr_balance.dtype}  and  of amount {amount.dtype}")
        
    def withdraw(self, amount) : 
        amount = self.num_check(amount)
        if(amount is not None):
            if amount > self.curr_balance:
                print(f"The amount you want to withdraw is bigger than your current balance of {self.curr_balance}")
            else : 
                self.curr_balance = self.curr_balance - amount
                print(f"Your withdrawal was successful! The current balance of {self.curr_user} is {self.curr_balance}")
```

```python
def main ():
    bank = Bank()
    while True:
        print("*" * 100)
        choice = input("Welcome to ALABAMA Bank\nWould you like to \n1)Register\n2)Login\n3)Exit\n")
        if choice == "3":
            break
        elif choice == "1":
            print("*" * 100)
            name = input("Please Enter your name : ")
            pin = input("Please Enter your pin : ")
            init = input("Please Enter your initial balance : ")
            bank.register(name, pin, init)
            
        elif choice == "2":
            print("*" * 100)
            while True:
                name = input("Please Enter your name : ")
                pin = input("Please Enter your pin : ")
                ans = bank.login(name, pin)
                if(ans):
                    break
                else:
                    print("Invalid username or pin...")
                    
            while True:
                print("*" * 50)
                action = input("Main Menu : \nWhat would you like to \n1)Check Balance\n2)Deposit\n3)Withdraw\n4)Exit\n")
                if action == "1":
                    bank.check_balance()
                    
                elif action == "2":
                    amount = input("Enter the amount you want to deposit : ")
                    init = bank.deposit(amount)
                    
                elif action == "3":
                    amount = input("Enter the amount you want to withdraw : ")
                    init = bank.withdraw(amount)
                        
                elif action == "4":
                    print(f"Ok sad to see you go your current balance is {init}, see you soon!")
                    break
        else:
            print("This transaction is invalid try again")
```

```python
main()
```
